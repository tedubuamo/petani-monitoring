```python
import pandas as pd
import re

# Membaca data dari file JSON
df_user = pd.read_json('user')

# Mendapatkan daftar email petani
userList = df_user['email_petani'].tolist()

# Input username dan password dari pengguna
username = input("username: ")
password = input("password: ")

# Regex pattern untuk validasi email
pattern = r'^[a-zA-Z0-9]+@petani\.com$'

if re.match(pattern, username):
    if username in userList:
        user_index = df_user.loc[df_user['email_petani'] == username].index[0]  # Mendapatkan indeks tunggal
        if df_user.at[user_index, 'password'] == password:  # Menggunakan .at untuk mendapatkan nilai tunggal
            print('login berhasil')
        else:
            print('Password salah')
    else:
        print('Email belum terdaftar')
else:
    print("Format email tidak valid")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Load Excel file
excel_file = 'your_excel_file.xlsx'

# List of sheet names
sheet_names = [f'GH{i}' for i in range(12, 0, -1)]

# Initialize an empty list to store DataFrames
dataframes = []

# Read each sheet into a DataFrame and store in the list
for sheet in sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet)
    
    # Convert 'Time' to datetime if necessary
    if 'Time' in df.columns:
        df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')
        
    dataframes.append(df)

# Combine all sheets into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

# Generate lag features for each column for 5-hour look back
hours_to_look_back = 5 * 60 // 30  # Assume data collected every 30 seconds
for i in range(1, hours_to_look_back + 1):
    combined_df[f'Humidity_lag_{i}'] = combined_df['Humidity'].shift(i)
    combined_df[f'Light_Intensity_lag_{i}'] = combined_df['Light Intensity'].shift(i)
    combined_df[f'Temperature_lag_{i}'] = combined_df['Temperature'].shift(i)
    combined_df[f'Soil_Moisture_lag_{i}'] = combined_df['Soil Moisture'].shift(i)

# Drop rows with NaN values generated by shifting
combined_df.dropna(inplace=True)

# Define features and targets
features = combined_df.drop(columns=['Time', 'Humidity', 'Light Intensity', 'Temperature', 'Soil Moisture'])
target_humidity = combined_df['Humidity']
target_light_intensity = combined_df['Light Intensity']
target_temperature = combined_df['Temperature']
target_soil_moisture = combined_df['Soil Moisture']

# Split data into train and test sets
X_train, X_test, y_train_humidity, y_test_humidity = train_test_split(features, target_humidity, test_size=0.2, random_state=42)
_, _, y_train_light, y_test_light = train_test_split(features, target_light_intensity, test_size=0.2, random_state=42)
_, _, y_train_temp, y_test_temp = train_test_split(features, target_temperature, test_size=0.2, random_state=42)
_, _, y_train_soil, y_test_soil = train_test_split(features, target_soil_moisture, test_size=0.2, random_state=42)

# Initialize linear regression model
linear_regression = LinearRegression()

# Train model for each target
linear_regression.fit(X_train, y_train_humidity)
y_pred_humidity = linear_regression.predict(X_test)
mse_humidity = mean_squared_error(y_test_humidity, y_pred_humidity)

linear_regression.fit(X_train, y_train_light)
y_pred_light = linear_regression.predict(X_test)
mse_light = mean_squared_error(y_test_light, y_pred_light)

linear_regression.fit(X_train, y_train_temp)
y_pred_temp = linear_regression.predict(X_test)
mse_temp = mean_squared_error(y_test_temp, y_pred_temp)

linear_regression.fit(X_train, y_train_soil)
y_pred_soil = linear_regression.predict(X_test)
mse_soil = mean_squared_error(y_test_soil, y_pred_soil)

# Print MSE for each prediction
print(f'MSE for Humidity: {mse_humidity}')
print(f'MSE for Light Intensity: {mse_light}')
print(f'MSE for Temperature: {mse_temp}')
print(f'MSE for Soil Moisture: {mse_soil}')
```
